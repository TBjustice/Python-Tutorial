# File Input
## How to read a Local File
To read a file, you need to open it(=get the access to the file), read it, and close it(=release the access to the file).<br>
To open a file, we can use ```open(filename)```.<br>
Read that with ```.read()```.<br>
Finally, close that with ```.close()```.
```python
file=open("numbers.txt")
text=file.read()
file.close()
```

````python
file=open("numbers.txt")
text=file.read()
file.close()
````

Let's check if the file is correctly read. Print first 30 charactors by running<br>
```python
print(text[0:30])
```

````python
print(text[0:30])
````

OK, the file is read correctly.<br>
As you can see, "numbers.txt" contains a list of floats each of which is separated by comma.<br>
How can we deal with this text? I mean, ```text``` is a mere sequence of charactors, but that is not what we want. We want to convert this string to list of numbers in order to do some calculation to it.<br>
<br>
Can we use ```float()```?<br>
No. ```float()``` can be used only for converting string to single float value.<br>
What you need to do first is to split this long sequence of text by comma.
```python
splitted = text.split(",")
```

````python
splitted = text.split(",")
````

Print the content and type of first 3 data in the ```splitted```. We can use for-loop for this.
```python
for element in splitted[0:3]:
    print(element)
    print(type(element))
```

````python
for element in splitted[0:3]:
    print(element)
    print(type(element))
````

Now you can see that ```splitted``` is a list that contains strings, each of which can be converted to float-numbers.<br>
To make a list of floats, run the following code.
```python
numbers=[]
for element in splitted:
    numbers.append(float(element))
print(numbers)
```

````python
numbers=[]
for element in splitted:
    numbers.append(float(element))
````

Print the content and type of first 3 data in the ```numbers```.
```python
for element in numbers[0:3]:
    print(element)
    print(type(element))
```

````python
for element in numbers[0:3]:
    print(element)
    print(type(element))
````

OK! Now you can get the list of numbers from the text file.<br>
<br>
As a review, I'll write the whole program below. Can you tell what happens at each line?
```python
file=open("numbers.txt")
text=file.read()
file.close()
splitted = text.split(",")

```

````python
file=open("numbers.txt")
text=file.read()
file.close()
splitted = text.split(",")
numbers=[]
for element in splitted:
    numbers.append(float(element))
````

## Show a Graph!
Let's make a graph about ```number```. First, import matplotlib.pyplot as plt<br>
```python

```

````python
import matplotlib.pyplot as plt
````

Create a figure of size 8 times 8, and create 2 axis.
```python
fig=plt.figure(figsize=(8,8))
ax1=fig.add_subplot(2,1,1)
ax2=fig.add_subplot(2,1,2)
```
Next, draw a histgram to ax1 by

```python

```
and draw line-graph to ax2 by

```python
ax2.plot(numbers)
```

Finally, show the graph by

```python
plt.show()
```

````python
fig=plt.figure(figsize=(8,8))
ax1=fig.add_subplot(2,1,1)
ax2=fig.add_subplot(2,1,2)
ax1.hist(numbers)
ax2.plot(numbers)
plt.show()
````
