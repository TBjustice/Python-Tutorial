# Matplotlib
## What is matplotlib?
Matplotlib is a tool for drawing graphs using python. Visualizing the data is very important in reaserch. The figure output from the library will decorate your paper, and catch readers eye!<br>
<br>
## Concept of Figures and Axes
Before you actually draw graphs, You must know about "Figures" and "Axes" in matplotlib.<br>

Axis is the canvas for drawing graphs, and figure wrapps one or more axes in it.<br>
<br>
Do you understand? If so, let's draw graphs!<br>
First, import matplotlib.pyplot (and abbreviate as plt)
```python
import matplotlib.pyplot as plt
```

```````python
import matplotlib.pyplot as plt
```````

Create a figure by ```plt.figure(configs)```. <br>
You can specify the size of figure by writing ```figsize=(width,height)``` in ```configs```.
This time, we'll create a figure that has a size of 8(width) times 4(height).<br>
<br>
Next, add axes to figure by calling ```add_subplot(NRow,NCol,id)```. <br>The place of axis is assigned by 3 numbers, ```NRow```,```NCol```, and ```id```. This means "Divide figure to have NRow-rows and NCol-columns. The axis is created in the ```id```th place."<br>
<br>
Finally, show the figure by calling ```plt.show()```.
```python
fig=plt.figure(figsize=(8,4))
ax1=fig.add_subplot(1,2,1)
ax2=fig.add_subplot(1,2,2)
plt.show()
```

```````python
fig=plt.figure(figsize=(8,4))
ax1=fig.add_subplot(1,2,1)
ax2=fig.add_subplot(1,2,2)
plt.show()
```````

The figure, whose size is 8 times 4, is created. Inside the figure are two axes, ```ax1``` and ```ax2.```. ```ax1``` is the left one and ```ax2``` is the right one.<br>
<br>
Next code generates 5 axes inside a figure. Check it out!
```python
fig=plt.figure(figsize=(12,8))
ax1=fig.add_subplot(2,3,1)
ax2=fig.add_subplot(2,3,2)
ax3=fig.add_subplot(2,3,3)
ax45=fig.add_subplot(2,3,(4,5))
ax6=fig.add_subplot(2,3,6)
plt.show()
```

```````python
fig=plt.figure(figsize=(12,8))
ax1=fig.add_subplot(2,3,1)
ax2=fig.add_subplot(2,3,2)
ax3=fig.add_subplot(2,3,3)
ax45=fig.add_subplot(2,3,(4,5))
ax6=fig.add_subplot(2,3,6)
plt.show()
```````

Five axes are created inside a single figure! Note that 4th and 5th place is connected and become single wide axis.<br>

## Add Title to Figure/Axis
You can set a title to figure by calling ```suptitle()```.<br>
Likewise, use ```set_title()``` to set a tile of axis.<br>
```python
fig=plt.figure(figsize=(8,4))

fig.suptitle("Title for figure")
ax1=fig.add_subplot(1,2,1)
ax1.set_title("Title for ax1")
ax2=fig.add_subplot(1,2,2)
ax2.set_title("Title for ax2")

plt.show()
```

```````python
fig=plt.figure(figsize=(8,4))

fig.suptitle("Title for figure")
ax1=fig.add_subplot(1,2,1)
ax1.set_title("Title for ax1")
ax2=fig.add_subplot(1,2,2)
ax2.set_title("Title for ax2")

plt.show()
```````

## Add Labal to X-Axis and Y-Axis
You can set label to two axes using ```set_xlabel()``` and ```set_ylabel()```. 
```python
fig=plt.figure(figsize=(4,4))

ax=fig.add_subplot(1,1,1)
ax.set_xlabel("x")
ax.set_ylabel("y")

plt.show()
```

```````python
fig=plt.figure(figsize=(4,4))

ax=fig.add_subplot(1,1,1)
ax.set_xlabel("time[min]")
ax.set_ylabel("weight[kg]")

plt.show()
```````

## Write Line Graphs
You can write line-graph useing ```.plot(Y)```, where ```Y``` is a list or numpy-array. See the sample code below.
```python
fig=plt.figure(figsize=(4,4))
ax1=fig.add_subplot(1,1,1)

ax1.plot([1,1,2,3,5,8,13])
ax1.set_xlabel("x(automatically created)")
ax1.set_ylabel("y")

plt.show()
```

```````python
fig=plt.figure(figsize=(4,4))
ax1=fig.add_subplot(1,1,1)

ax1.plot([1,1,2,3,5,8,13])
ax1.set_xlabel("x(automatically created)")
ax1.set_ylabel("y")

plt.show()
```````

You can see that the x-axis is automatically created.<br>
<br>
When you want to assign the value of x-axis, just write ```ax1.plot(X, Y)```.<br>

## Add Legend to Graph
You can designate a label to each graph by writing ```label=""```.<br>
After that, call ```.legend()``` to show a legend.
```python
fig=plt.figure(figsize=(8,4))
ax1=fig.add_subplot(1,1,1)

ax1.plot(x, y1, label="y=sin(x)")
ax1.plot(x, y2, label="y=cos(x)")
ax1.set_xlabel("x")
ax1.set_ylabel("y")
ax1.legend()

plt.show()
```

```````python
fig=plt.figure(figsize=(8,4))
ax1=fig.add_subplot(1,1,1)

ax1.plot(x, y1, label="y=sin(x)")
ax1.plot(x, y2, label="y=cos(x)")
ax1.set_xlabel("x")
ax1.set_ylabel("y")
ax1.legend()

plt.show()
```````

## Histgram
To draw histgram, call ```.hist()``` to axis.<br>
```np.random.normal(0, 1, 100)``` will create a numpy-array of 100 elements, whose average is 0 and standard division is 1.
```python
data=np.random.normal(0, 1, 100)

fig=plt.figure(figsize=(8,4))
ax1=fig.add_subplot(1,1,1)

ax1.hist(data)
ax1.set_title("np.random.normal(100)")

plt.show()
```

```````python
data=np.random.normal(0, 1, 100)

fig=plt.figure(figsize=(8,4))
ax1=fig.add_subplot(1,1,1)

ax1.hist(data)
ax1.set_title("np.random.normal(100)")

plt.show()
```````

## Scatter plot
To draw scatter plots, call .scatter(x,y) to axis.<br>
```np.random.rand(100)``` creates numpy-array of 100 elements, whose value is randomly chosen from 0 to 1.
```python
x=np.random.rand(100)*5
y=x+np.random.normal(0, 1, 100)

fig=plt.figure(figsize=(8,4))
ax1=fig.add_subplot(1,2,1)
ax2=fig.add_subplot(1,2,2)

ax1.scatter(x, y)
ax1.set_xlabel("x")
ax1.set_ylabel("y")

ax2.scatter(x, y, marker="*")
ax2.set_xlabel("x")
ax2.set_ylabel("y")

plt.show()
```

```````python
x=np.random.rand(100)*5
y=x+np.random.normal(0, 1, 100)

fig=plt.figure(figsize=(8,4))
ax1=fig.add_subplot(1,2,1)
ax2=fig.add_subplot(1,2,2)

ax1.scatter(x, y)
ax1.set_xlabel("x")
ax1.set_ylabel("y")

ax2.scatter(x, y, marker="*")
ax2.set_xlabel("x")
ax2.set_ylabel("y")

plt.show()
```````

You can specify the shape of marker by writing ```marker=""```<br>
`````` can be ```o```(circle), ```*```(star), ```^```(triangle), and many more. Check all the keywords by looking official website for matplotlib.<br>

## TwinX
When you want to place two axes in the same place(drawing climograph, for example), you can use ```ax.twinx()```. Check the sample code below.
```python
month=np.arange(12)+1
temp=np.array([1, 0, 8, 15, 21, 25, 28, 30, 25, 21, 12, 5])
rain=np.array([50, 70, 100, 110, 115, 160, 150, 135, 175, 121, 60, 55])

fig=plt.figure(figsize=(8,8))
ax1=fig.add_subplot(1,1,1)
ax2=ax1.twinx()

ax1.plot(month, temp, color="#ff0000")
ax1.set_ylim((-10, 40))
ax1.set_ylabel("Temperature")

ax2.bar(month, rain, color="#6688ff")
ax2.set_ylim((0, 400))
ax2.set_ylabel("Precipitation")

plt.show()
```

```````python
month=np.arange(12)+1
temp=np.array([1, 0, 8, 15, 21, 25, 28, 30, 25, 21, 12, 5])
rain=np.array([50, 70, 100, 110, 115, 160, 150, 135, 175, 121, 60, 55])

fig=plt.figure(figsize=(8,8))
ax1=fig.add_subplot(1,1,1)
ax2=ax1.twinx()

ax1.plot(month, temp, color="#ff0000")
ax1.set_ylim((-10, 40))
ax1.set_ylabel("Temperature")

ax2.bar(month, rain, color="#6688ff")
ax2.set_ylim((0, 400))
ax2.set_ylabel("Precipitation")

plt.show()
```````

```ax1``` and ```ax2``` share X-axis, but they have different Y-axis. This is how ```twinx()``` works.
